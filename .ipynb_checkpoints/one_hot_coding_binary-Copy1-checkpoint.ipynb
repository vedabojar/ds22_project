```python
import pandas as pd

def one_hot_encoding_binary(dataset, orginal_column, new_column_name, file_location, file_column="Name", separator=",", process_all_names=False):
    
    split_values = dataset[orginal_column]

    # 2. Selecting only the first or all names based on the 'process_all_names' argument
    if process_all_names:
        values_to_process = split_values
    else:
        values_to_process = split_values.str[0]

    # 3. Creating the new column with binary values
    new_column_values = []
    for value in values_to_process:
        if isinstance(value, str):  # added check for string type
            directors = value.split(',')
            flag = False
            for director in directors:
                if director.strip() in pd.read_csv(file_location)[file_column].values:
                    flag = True
                    break
            new_column_values.append(int(flag))
        else:
            new_column_values.append(0)
    dataset[new_column_name] = new_column_values

    # 4. Dropping the original column
    dataset.drop(columns=[orginal_column], inplace=True)

    # 5. Check the updated table
    #dataset.head(2)
    return dataset
```

```python
dataset_prosscessed = one_hot_encoding_binary(dataset_raw, "Directors", "Top_50_Director", "C:/Users/admin1/Documents/GitHub/ds22_project/data/top_1000_actors.csv",  process_all_names=True)
```
