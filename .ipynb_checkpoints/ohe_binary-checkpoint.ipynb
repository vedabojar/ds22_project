```python
import pandas as pd
```

```python
dataset_raw = pd.read_csv('C:/Users/admin1/Documents/GitHub/ds22_project/data/movies.csv')
dataset_raw.head(2)
```

```python
def one_hot_coding_binary(dataset, original_column, new_column, file_column, file_location, seperator = ", "):
    
    # 1. Only keeping the first director
    dataset[new_column] = dataset[original_column].str.split(seperator).str[0]

    # 2. Dropping the original column
    dataset.drop(columns=[original_column], inplace=True)

    # 1. Importing top directors dataset
    dataset_directors = pd.read_csv(file_location)

    # 2. If Director is in dataset_directors, replace name with 1
    # else replace name with 0
    replace = dataset[new_column].isin(dataset_directors[file_column])
    dataset[new_column] = replace.astype(int)
    
    return dataset

dataset_new = one_hot_coding_binary(dataset_raw, "Directors", "Top_50_Director", "Name", "C:/Users/admin1/Documents/GitHub/ds22_project/data/top_50_directors.csv")
```
